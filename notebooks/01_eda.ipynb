```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/kaggle/input/mtsamples/mtsamples.csv')
```

```python
# Strip Whitespace

df['medical_specialty'] = df['medical_specialty'].str.strip()
print("Whitespace stripped from medical_specialty")
```

```python
# Specialty Distribution

specialty_counts = df['medical_specialty'].value_counts()

print(f"Total unique specialties: {len(specialty_counts)}")
print("\nTop 15 Specialties:")
print(specialty_counts.head(15))
```

```python
# Plot Distribution

plt.figure(figsize=(14, 6))
specialty_counts.head(15).plot(kind='bar', color='steelblue')
plt.title('Top 15 Medical Specialty Distribution', fontsize=14)
plt.xlabel('Medical Specialty')
plt.ylabel('Number of Samples')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('specialty_distribution.png', dpi=150)
plt.show()
```

```python
# Text Length Analysis

df['text_length'] = df['transcription'].apply(
    lambda x: len(str(x).split()) if pd.notnull(x) else 0
)

print("Transcription Length Statistics (words):")
print("-" * 40)
print(f"Average : {df['text_length'].mean():.0f} words")
print(f"Minimum : {df['text_length'].min()} words")
print(f"Maximum : {df['text_length'].max()} words")
print(f"Median  : {df['text_length'].median():.0f} words")

above_512 = (df['text_length'] > 450).sum()
print(f"\nTranscriptions above 450 words: {above_512} ({above_512/len(df)*100:.1f}%)")
```

```python
# Text Length Distribution

plt.figure(figsize=(12, 5))
df['text_length'].hist(bins=50, color='steelblue', edgecolor='white')
plt.axvline(x=512, color='red', linestyle='--', label='DistilBERT 512 token limit')
plt.title('Transcription Length Distribution')
plt.xlabel('Word Count')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Summary

print("EDA Summary")
print("=" * 40)
print(f"Total samples         : {len(df)}")
print(f"Total specialties     : {df['medical_specialty'].nunique()}")
print(f"Null transcriptions   : {df['transcription'].isnull().sum()}")
print(f"Avg text length       : {df['text_length'].mean():.0f} words")
print(f"Texts above 450 words : {above_512} ({above_512/len(df)*100:.1f}%)")
print("\nNext step: 02_preprocessing.ipynb")
```
